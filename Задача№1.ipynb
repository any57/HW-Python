```
def get_info_film() -> Iterable[dict]:
  page_number: int = 1
  soup = BeautifulSoup(requests.get(url='https://www.kinoafisha.info/rating/movies/?page=1#').text,"lxml").find_all('div', attrs = {'class':"films_right"})
  while soup:
    url_page = 'https://www.kinoafisha.info/rating/movies/?page={}#'.format(page_number)
    soup = BeautifulSoup(requests.get(url=url_page).text,"lxml").find_all('div', attrs = {'class':"films_right"})
    page_number += 1
    for i in soup:
      dict_info: dict = {}
      dict_info['name'] = i.find('a', attrs = {'class':'films_name ref'}).text
      dict_info['kp_rating'] = i.find('span', attrs = {'class':'rating_num'}).text
      dict_info['year'] = i.find('span', attrs = {'class':'films_info'}).text[:4]
      dict_info['url'] = 'https://www.kinoafisha.info/' + i.find('a', attrs = {'class':'films_name ref'})['href']
      dict_info['_utc_timestamp'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
      time.sleep(0.5)
      yield dict_info
```

```
def get_info_film(time_for_sleep: Union[int, float], number: int) -> Iterable[dict]:
  if (time_for_sleep < 0):
    return "Отрицательное время паузы"
  if (number<0):
    return "Отрицательное количество фильмов"
  page_number = 1
  soup = BeautifulSoup(requests.get(url='https://www.kinoafisha.info/rating/movies/?page=1#').text,"lxml").find_all('div', attrs = {'class':"films_right"})
  while soup:
    url_page = 'https://www.kinoafisha.info/rating/movies/?page={}#'.format(page_number)
    soup = BeautifulSoup(requests.get(url=url_page).text,"lxml").find_all('div', attrs = {'class':"films_right"})
    page_number += 1
    for i in soup:
      if number == 0:
        return
      number -= 1
      dict_info = {}
      dict_info['name'] = i.find('a', attrs = {'class':'films_name ref'}).text
      dict_info['kp_rating'] = i.find('span', attrs = {'class':'rating_num'}).text
      dict_info['year'] = i.find('span', attrs = {'class':'films_info'}).text[:4]
      dict_info['url'] = 'https://www.kinoafisha.info/' + i.find('a', attrs = {'class':'films_name ref'})['href']
      dict_info['_utc_timestamp'] = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
      time.sleep(time_for_sleep)
      yield dict_info
```

```
# Функция должна иметь следующую сигнатуру (набор и тип аргументов)
def dump(film_list: List[Dict[str, Any]], 
         path: str, 
         filetype: str = 'json', 
         order_by: Union[None, str] = None, # поле словарей, по которому будем сортировать
         ascending: bool = True) -> None:
    with open(path + "." + filetype, 'w') as f:
      if order_by:
        json.dump(sorted(film_list, key=lambda x: x.get(order_by),reverse = ascending), f,ensure_ascii=False)
      else:
        json.dump(film_list, f,ensure_ascii=False)
# Пример вызова:
my_list = list_of_dict_info

# Записываем csv файл, отсортированный по убыванию времени выгрузки.
dump(list_of_dict_info, 'myfile', order_by = "kp_rating",filetype='json', ascending=False)
```
